### Distance and speed conversions

**Q1.** You have been informed that `1` mile is equivalent to `1.61` kilometres.

Based on the given facts create a function called `km` with a single parameter called `miles`, (which you can assume will always be a valid number) and returns the equivalent number of kilometres.



```python

def km(miles):
    kilometers = miles * 1.61
    return kilometers

```

**Q3.** Create a function called `over_kmph_limit` which takes two parameters:
- `mph_speed`, representing the speed of a car in miles per hour

```python

def over_kmph_limit(mph_speed, kmph_limit):
    if mph_speed * 1.61 > kmph_limit:
        return True
    else:
        return False

```

**Q5.** Create a function called `adults` with a single parameter called `members`, which takes a list of tuples in the format given above (`name`, `age`, `gender`), and returns a list of names (the names only) of those who are aged 18 or over.


- Hint: Consider using a `for` loop to iterate through values in the list
- Use an `if condition` within the `for` loop to check for age values: 18 or over 
- Also make sure to create an `empty list` to append the people who are aged 18 or over



```python

def adults(members):
    adults_list = [] 

    for person in members:
        if person[1] >= 18: 
            adults_list.append(person[0])  
    
    return adults_list  


```

**Q6.** Create a function called `average_age` which takes two parameters:

- `members`, a list of tuples in the same format as `people` above 
- `gender`, a text string such as (but not limited to) `'M'` or `'F'`.

It should return the average age of entries in `members` whose gender matches the argument given for `gender`.

If no matches are found, return the string `'No matches found.'` 

As before, ensure this string is formatted exactly as given here.

```python

def average_age(members, gender):
    total_age = 0
    count = 0

    for person in members:
        if person[2] == gender:  
            total_age += person[1]  
            count += 1  

    if count == 0:
        return 'No matches found.'  
    
    avg_age = total_age / count  
    return avg_age  

```

```python
average_age(people, 'M')
```

```python
average_age(people, 'Z')
```

**Q7.** Create a function called `member_details`, with two parameters:

- `entries`, a list of tuples, in the same format as the `people` list above
- `name`, a string, representing the name of one of the entries

The function should return a dictionary with two keys, `age` and `gender`, with the values being those from the tuple which contains the name passed as the second argument to the function. If the name isn't found in the list of tuples, return [`None`](https://www.w3schools.com/python/ref_keyword_none.asp).


```python

def member_details(entries, name):
    for entry in entries:
        if entry[0] == name:  
            
            details = {'age': entry[1], 'gender': entry[2]}
            return details  
    
    return None  

```

```python
ahmed_details = member_details(people, 'Ahmed')
ahmed_details
```

**Q8.** Create a function called `check_group` which has two arguments:

- `members` a list of names representing existing members (just a list, not tuples as previously)
- `group` a list of names representing visitors (who may or may not be members)

The function should return a string in the `f-string` format `f'Member present: {person}.'` (ensure it matches this exactly), where `{person}` is replaced by the first name from the visitors list to be identified as a member.

If no member is found, the function should return a string `'No members.'` (ensure it matches this exactly)



```python

def check_group(members, group):
    for person in group:
        if person in members:
            return f'Member present: {person}.' 

    return 'No members.' 

```
